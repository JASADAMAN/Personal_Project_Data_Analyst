we use percentile  with IQR to remove outlier

```python
# Finding the IQR 
# ride_time_min
time_min_p25 = clean_merge['ride_time_min'].quantile(0.25)
time_min_p75 = clean_merge['ride_time_min'].quantile(0.75)

# age
age_p25 = clean_merge['Age'].quantile(0.25)
age_p75 = clean_merge['Age'].quantile(0.75)
```

```python
# create iqr
#ride_time_min
t_iqr = time_min_p75 - time_min_p25

#age
a_iqr = age_p75 - age_p25
```

```python
# find lower / upper limit
#ride_time_min
t_upper_limit = time_min_p75 + 1.5 * t_iqr
t_lower_limit = time_min_p25 - 1.5 * t_iqr
# age
a_upper_limit = age_p75 + 1.5 * a_iqr
a_lower_limit = age_p25 - 1.5 * a_iqr
```

```python
# Finding outliers
#ride
clean_merge[clean_merge['ride_time_min'] > t_upper_limit]
clean_merge[clean_merge['ride_time_min'] < t_lower_limit]
```

```python
#age
clean_merge[clean_merge['Age'] > a_upper_limit]
clean_merge[clean_merge['Age'] < a_lower_limit]
```

```python
# Trim outlier
clean_otlier = clean_merge[clean_merge['ride_time_min'] < t_upper_limit ] 
clean_outlier = clean_otlier[clean_otlier['ride_time_min'] > t_lower_limit ]
clean_outlier.shape
```

```python
clean_outlier.describe()
```

```python
fig = plt.figure(figsize =(10, 7))
 
# Creating plot
sns.boxplot(x=clean_outlier["ride_time_min"])
# show plot
plt.show()
```

```python
# Trim outlier
#age
clean_outlier_final = clean_outlier[clean_outlier['Age'] < a_upper_limit ]  
clean_outlier_final.shape
```

```python
fig = plt.figure(figsize =(10, 7))
 
# Creating plot
sns.boxplot(x=clean_outlier_final["Age"])
# show plot
plt.show()
```
